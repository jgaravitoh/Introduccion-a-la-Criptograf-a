```python
def VignereCipher(msg, key, t, encrypt):

  alphabet = list(map(chr, range(65, 91)))  #alphabet declaration
  #making msg and key UPPERCASE
  msg = msg.upper()
  key = key.upper()
  #making msg and key not kaboom kaboom the code (Delete any unwanted character)
  for char in msg:
    if char not in alphabet:
      msg = msg.replace(char,"")
  for char in key:
    if char not in alphabet:
      key = key.replace(char,"")
  # making key the same size as msg
  while(len(key)<len(msg)):
    key += key
  key = key[0:len(msg)]

  # encrypt adding up the index of character, and decrypt substracting the index of character
  ret_msg = ""
  counter = 0
  if encrypt:
    for char_msg, char_key in zip(msg,key):
      ret_msg += alphabet[(alphabet.index(char_msg) + alphabet.index(char_key)) % len(alphabet)]
      counter += 1
      if counter == t:
        ret_msg += " "
        counter = 0
  else:
    for char_msg, char_key in zip(msg,key):
      ret_msg += alphabet[(alphabet.index(char_msg) - alphabet.index(char_key)) % len(alphabet)]
      counter += 1
      if counter == t:
        ret_msg += " "
        counter = 0
  return ret_msg
```

```python
# TO ENCRYPT USE True AS THE LAST PARAMETER, TO DECRYPT USE False
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Vigènere Cipher ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n");
print(" TO ENCRYPT USE True AS THE LAST PARAMETER, TO DECRYPT USE False ");
#plain_text = "TO BE OR NOT TO BE THAT IS THE QUESTION"
plain_text = input("Please enter the message: ");
#key = "RELATIONS"
key =  input("Please enter the key: ");
#t = 2;
t = int(input("Please enter the t parameter: "));
encrypted = VignereCipher(plain_text, key,t, True);
decrypted = VignereCipher(encrypted, key,t, False);
print("PlainText: "+plain_text+ "  |  Key: " + key);
print("Encrypted: "+encrypted);
print("Decrypted: "+decrypted);
```
