```python
def OTP(message, key):
  # Check lenght of message and key
  if(len(message)!=len(key)):
    raise Exception("ERROR: Key and message must be the same lenght")
  ret_message = ""
  # Python's xor can be done with the ASCII code
  for char_msg,char_key in zip(message,key):
    ret_message += chr(((ord(char_msg) - ord(" ")) ^ ord(char_key))%96 + ord(" "))
  return ret_message
```

```python
print("~~~~~~~~~~~ One Time Pad ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")
plain_text = "Hello World"
key = "86f3owf5h7w"
encrypted = OTP(plain_text, key)
decrypted = OTP(encrypted, key)
print("PlainText: "+plain_text+ "\nKey: " + key)
print("Encrypted: "+encrypted)
print("Decrypted: "+decrypted)
```
