```python
import math
import decimal
def calculate_possible_layouts_lower(m = 26, n = 100):
    combinations_n_m = math.comb(n, m);
    factorial_m = math.factorial(m);
    power_term = m**(n - m);
    print("   combinations(n,m)",combinations_n_m);
    print("   m!:", factorial_m);
    print("   m^(n-m):", power_term);
    return combinations_n_m * factorial_m * power_term;

def calculate_possible_layouts_upper(m = 26, n = 100):
    return m**n;

def calculate_possible_layouts_estimate(m = 26, n = 100):
    return 2 ** (4.7 * n);
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
print("Lower bound:")
possible_layouts = calculate_possible_layouts_lower();
print(" Number of possible layouts:", possible_layouts)
print(" Order of possible layouts:",'{:.2e}'.format(possible_layouts))


print("\nUpper bound:")
possible_layouts = calculate_possible_layouts_upper();
print(" Number of possible layouts:", possible_layouts)
print(" Order of possible layouts:",'{:.2e}'.format(possible_layouts))

print("\nEstimate bound:")
possible_layouts = calculate_possible_layouts_estimate();
print(" Number of possible layouts:", possible_layouts)
print(" Order of possible layouts:",'{:.2e}'.format(possible_layouts))
```

```python
def homophonic_cipher(msg, layout, encrypt_bool):
  import random;
  msg = ''.join(char.upper() for char in msg if (char.isalpha() and encrypt_bool) or ((char.isnumeric() and not encrypt_bool) or (char == " " and not encrypt_bool)));
  if(len(msg)==0): raise(Exception("ERROR: message cannot be of lenght 0"));
  ret_msg = "";

  if encrypt_bool:
    for char in msg:
      letter_index = ord(char) - ord('A');
      random_number = random.choice(list(layout[letter_index]));
      #letter = chr(ord('A') + random_index);
      ret_msg += str(random_number) + " ";
  else:
    msg = list(map(int, msg.split()))
    for i in range(len(msg)):
      for j in range(len(layout)):
        if msg[i] in layout[j]:
          ret_msg += chr(ord('A') + j)+" ";
  return ret_msg
```

```python
print("~~~~~~~~~~~ Homophonic Cipher ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")
plain_text = "Estoy en villeta";
encrypted = homophonic_cipher(plain_text,layout,True);
decrypted = homophonic_cipher(encrypted,layout,False);

print("Layout: m = "+ str(sum(len(subset) for subset in layout)), ", n = "+ str(len(layout)))
print("PlainText: "+plain_text)
print("Encrypted: "+encrypted)
print("Decrypted: "+decrypted)
```
