```python
import string;
from collections import OrderedDict
```

```python
# OrderedDict is a dictionary subclass that remembers the order of the keys that
# were inserted first. Since there can't be duplicate keys this method will
# return the string after removing the duplicate characters.
# https://www.studytonight.com/python-programs/remove-all-duplicates-from-a-given-string-in-python

def FindPosition(matrix, letter):
    for i in range(len(matrix)):
      for j in range(len(matrix[i])):
        for z in range(len(matrix[i][j])):
          if matrix[i][j][z] == letter:
            return i, j

def RemoveDuplicates(text):
    return "".join(OrderedDict.fromkeys(text))

def MakeKeyWordMatrix(keyword,cols,rows):
  alphabet = string.ascii_uppercase;
  if (cols*rows > len(alphabet)):
    raise Exception("ERROR: It's not possible to use a matrix greater than the size of the alphabet")
  keyword = keyword.upper();
  char_matrix = [];
  chars_to_add = RemoveDuplicates(((keyword + alphabet).replace("J","I")).replace(" ",""));
  chars_index = 0;
  for i in range(cols):
    char_matrix.append([]);
    for j in range(rows):
      if chars_to_add[chars_index] == "I":
        char_matrix[i].append([]);
        char_matrix[i][j].append("I");
        char_matrix[i][j].append("J");
      else:
        char_matrix[i].append(chars_to_add[chars_index]);
      chars_index +=1;
  return char_matrix;

def PrintMatrix(matrix):
  for row in matrix:
    print(row);

keyword = "yoanpiz"
cols = 5;
rows = 5;
#print(FindPosition(MakeKeyWordMatrix(keyword,cols,rows),"I"))
```

```python
def PlayFair(msg, keyword, encrypt_bool, print_bool):
  cols = 5;
  rows = 5;
  keyword_matrix = MakeKeyWordMatrix(keyword,cols,rows);
  if print_bool: PrintMatrix(keyword_matrix);
  msg = msg.upper().replace(" ","");
  new_msg = ""

  if len(msg)>=2:
    for i in range(len(msg)-1):
      new_msg += msg[i];
      if i%2 == 0:
        if msg[i] == msg[i+1]:
          new_msg+="X";
    new_msg += msg[len(msg)-1];
    msg=new_msg;


  ret_msg = "";
  shifting = -1;
  if encrypt_bool == True:
    shifting = 1;
    if len(msg)%2 != 0:
      msg+="X"
  i = 0;
  while i < (len(msg)):
    char1_pos = FindPosition(keyword_matrix,msg[i]); # access to  char_pos[0] and char_pos[1]
    char2_pos = FindPosition(keyword_matrix,msg[i+1]);
    i += 2
    # cases
    if(char1_pos[1] == char2_pos[1]):
      ret_msg += keyword_matrix[(char1_pos[0]+shifting)%(rows)][(char1_pos[1])][0];
      ret_msg += keyword_matrix[(char2_pos[0]+shifting)%(rows)][(char2_pos[1])][0];
    elif(char1_pos[0] == char2_pos[0]):
      ret_msg += keyword_matrix[char1_pos[0]][((char1_pos[1]+shifting)%(cols))][0];
      ret_msg += keyword_matrix[char2_pos[0]][((char2_pos[1]+shifting)%(cols))][0];
    else: # not matching row or column
      ret_msg += keyword_matrix[char1_pos[0]][(char2_pos[1])][0];
      ret_msg += keyword_matrix[char2_pos[0]][(char1_pos[1])][0];
  return ret_msg

```

```python
print("~~~~~~~~~~~ PlayFair Cipher ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")
plain_text = "THIS SECRET MESSAGE IS ENCRYPTED";
keyword = "YOANPIZ";
print("\n Matrix: \n")
encrypted = PlayFair(plain_text, keyword, True, True);
decrypted = PlayFair(encrypted, keyword, False, False);
print("\n")
print("PlainText: "+plain_text+ "\nKeyword: " + keyword)
print("Encrypted: "+encrypted)
print("Decrypted: "+decrypted)
```
