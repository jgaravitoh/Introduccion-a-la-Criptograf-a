```python
import numpy as np;
import string;
import sympy;
```

```python
def HillCipher(msg, key, encrypt_bool, filler = "x"):
  #msg = msg.lower().replace(" ","");
  #alphabet = list((string.ascii_lowercase));
  alphabet = list((string.printable));
  ret_msg = "";
  m_vector = []
  t = len(key);
  key = sympy.Matrix(key);

  # Verifications of key
  if key.det() == 0:
    raise("ERROR: Key's (matrix) determinant cannot be 0");
  if sympy.gcd(key.det(), t) != 1:
    raise("ERROR: Key's (matrix) determinant and 1 must be coprime integers");

  key = key.inv_mod(len(alphabet)) if not encrypt_bool else key;

  # FILLER
  if len(msg)%t > 0:
    for i in range(len(msg)%t):
      msg+=filler;
  # CIPHER
  for i in range(len(msg)):
    if(msg[i] not in alphabet):
      raise Exception("ERROR: Only letters in the alphabet are allowed!");
    m_vector.append(alphabet.index(msg[i]))
    if i%t-1 == 0:
        if len(msg)>1 and i == 0:
          continue
        m_vector = np.dot(m_vector, key)%len(alphabet);
        for j in range(t):
          ret_msg += alphabet[m_vector[j]];
        m_vector = [];
  return ret_msg;
```

```python
print("~~~~~~~~~~~ Hill Cipher ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")
plain_text = "number theory is easy123!_)@()??/\\";
key = [[11,8],[3,7]];
encrypted = HillCipher(plain_text,key,True);
decrypted = HillCipher(encrypted,key,False);
print("PlainText: "+plain_text+ "\nKey: " +str(key))
print("Encrypted: "+encrypted)
print("Decrypted: "+decrypted)
```
